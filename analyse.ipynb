```python
# convert time to unix time in TimeSunRise and TimeSunSet using pd.Timestamp
# lambda x where type(x) is datetime.time
df['TimeSunRise'] = df['TimeSunRise'].apply(lambda x: str(pd.Timestamp(x.strftime('%H:%M:%S')).timestamp()))
```

```python
# here x is string
df['TimeSunSet'] = df['TimeSunSet'].apply(lambda x: str(pd.Timestamp(x).timestamp()))
# df['TimeSunSet'] = df['TimeSunSet'].apply(lambda x: pd.Timestamp(x.strftime('%H:%M:%S')).timestamp())

```

```python
# use StandardScaler to scale the data
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler object
scaler = StandardScaler()

# Standardize the numeric features
numeric_features = df.select_dtypes(include=[np.number])
scaled_features = scaler.fit_transform(numeric_features)

# Create a DataFrame to store the scaled numeric features
scaled_df = pd.DataFrame(data=scaled_features, columns=numeric_features.columns)
scaled_df
```

```python
# Perform PCA (Principal Component Analysis)
from sklearn.decomposition import PCA

# Decide the number of components (e.g., 2 components)
n_components = 2

# Fit PCA on the standardized numeric features
pca = PCA(n_components=n_components)
principal_components = pca.fit_transform(scaled_features)

# Create a DataFrame to store the principal components
pca_df = pd.DataFrame(data=principal_components, columns=[f'PC{i+1}' for i in range(n_components)])
print("Explained variance ratio:", pca.explained_variance_ratio_)

# Combine PCA results with the original DataFrame for modeling
df = pd.concat([df.reset_index(drop=True), pca_df.reset_index(drop=True)], axis=1)
```

```python
df
```

```python
heatmap = sns.heatmap(df.corr(), annot=True)
```

```python
# use random forest to predict the data
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split


X = df.drop('Radiation', axis=1)
y = df['Radiation']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=5)
```

```python
model = RandomForestRegressor()
model.fit(X_train, y_train)
```

```python

model.score(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, root_mean_squared_error, mean_absolute_error, r2_score

print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error:", root_mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```
